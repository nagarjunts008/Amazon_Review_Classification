```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('amazon_review.csv')

print(df.isnull().sum(axis = 0))

df1 = df[df['text'].isna()]
df2 = df[df['review_id'].isna()]
df3 = df[df['review_score'].isna()]
df4 = df[df['product_category'].isna()]

a=df1.append(df2)
b=a.append(df3)
c=b.append(df4)

null_index = c.index.tolist()

print('\n', null_index)
print('\n', len(null_index))
```

```python
df1 = df[df.duplicated(subset=['text'])]
duplicated_index = df1.index.tolist()

print('\n', duplicated_index)
print('\n', len(duplicated_index))
```

```python
df = pd.read_csv('amazon_review.csv')
negative_score_index = df[ df['review_score'] == -1 ].index.tolist()

print('\n', negative_score_index)
print('\n', len(negative_score_index))
```

```python
df['Exclude'] = '0'
df['Reason'] = ''
df
```

```python
for i in null_index:
    df['Exclude'][i] = 1
    df['Reason'][i] = "Missing values"

for i in duplicated_index:
    df['Exclude'][i] = 1
    df['Reason'][i] = "Duplicated"

for i in negative_score_index:
    df['Exclude'][i] = 1
    df['Reason'][i] = "Negative score value"

df.to_csv('excluded_data.csv')
df
```
